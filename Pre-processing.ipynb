```python
data2 = data.copy()

for column in data2.columns:
    # Calculate the mode of the column
    mode = data2[column].mode()  # Use iloc[0] to get the first mode (in case of multiple modes)

    # Replace missing values in the column with the mode
    data2[column].fillna(mode, inplace=True)
```
